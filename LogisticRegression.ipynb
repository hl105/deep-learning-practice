**Logistic Regression**

Logistic regression is one of the most commonly used classification algorithm. Below are some questions I had while learning this classification technique, and summarized answers (that I looked up) to those questions.


---



**Q: Regression? I thought regression was used to predict numeric values?**

**A:** Review: regression models are used to predict numeric values, while classification models are used to map data to discrete class labels.


The **regression** part of logistic regression comes from the following process:

*   The model regresses the **probability** of the sample having each label.

Then assigns the class with the highest probability as the sample's class

---

**Q: What's the logistic part?**

**A:** We use a sigmoid function (a type of logistic function) to get the probability and get the decision boundary for classifciation.

---
**Q:Linear regression vs. logistic regression?**

**A:**

The goal of linear regression is to fit a linear equation to data by minimizing SSE. Optimization is just getting the one and only global minimum of the loss funcion -- LSE -- in convex graph form.

The goal of logistic regression is to model the probabilty of the input belonging in a certain class. So we can't use LSE anymore since its hard to find a global minimum. Instead, we use the MAximum Likelyhood Estimation.

---

**Q: Generative vs discriminative model?**

**A:**

Generative model: P(x,y:Θ) - have to get joint probability

Discriminative model: P(y|x:Θ) - have to only get y's prob when input is x

Logistic regression is a discriminative model

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
df = load_breast_cancer(as_frame = True).frame
df.head()
```

**Debugging:**

At first, I **did NOT scale the features** since I was not dealing with distances or needed the features to be on a simliar scale. But then I realized that I needed to adjust the max_iter parameter from the default 100 to 5000 in the **LogisticRegression()** function. I learned that scaling data speeds up convergence, so we won't need to adjust the number of iterations.

```python
# Choose independent(feature) variables and dependent(target) variables
X = df.drop('target', axis = 1) #axis = 1 means column
y = df['target']

#split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 20)
```

```python
# feature scaling
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.fit_transform(X_test)
```

```python
# define model
logReg = LogisticRegression()

# fit model
logReg.fit(X_train, y_train)

# predict
y_pred = logReg.predict(X_test)
```

```python
# confusion matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(accuracy_score(y_test,y_pred))
```
